    <i>To check and see if two graphs that may or may not look different are actually the same?</i> Essentially, if we have two graphs, do they with certainty have a One to One mapping of the vertices in one graph to the vertices of the other such that adjacency is preserved? </p>

<pre>

    |E(G2)| = |E(G1)|
</pre>
<p>Means we have two isomorphic graphs.</p>

# P versus NP problem.

<p>To try to keep this straight forward initially, P is a set of relatively easy problems, and NP is a set of of what seems to be very hard problems. So, P = NP would imply that hard problems would actually have relatively easy solutions. But this has more complexity than that. So much that Clay Mathematics Institute will give you one million dollars for proving or disproving this theory!</p>

<p>The problem is unsolved in computer science. It creates the question if the solution of a problem can be quickly verified, can it also be solved quickly?</p>

<p>The term "quickly" is referred to the use of an algorithm, which provides a solution or outcome in polynomial time (<b>P</b>). A polynomial is an expression variables, terms, exponents and constants. Polynomial time is define by the running time is upper bounded by a polynomial expression which is the size of the input for the algorithm. Noted: T(n) = O(nk) for some positive constant k. For example, basic arithmetic operations (Addition, subtraction, etc) can be performed in polynomial time.</p>
<p>Sometimes, there is no known way to find an answer <b>but, if we are provided information showing the answer</b>, it is possible to verify the answer quickly. (<b>NP</b>) NP, which stands for nondeterministic polynomial, is the category these questions fall into if they can be answered/verified in polynomial time.</p>

<p>To conclude this section. If we are asked does P equal NP, we are asking if the solution to a problem can be verified in polynomial time, can it also be found in polynomial time.</p>

### Isomorphism Algorithm

***

<p>Now that we have a better understanding of graphs and how graphs can be used to represent data structures, let's use an algorithm to see if two graphs are isomorphic or not and further assess the graph isomorphism problem. Before diving into this, it is important to note again that the graph isomorphism problem is not yet solved and the below algorithm is highly inefficient. But, as inefficient as it may be, it will help us better understand the complexity of this topic.</p>

<p>The <i>bruteForseIso</i> function below will take in two adjacency matrices and will return us either a true or false result which states if these graphs are isomorphic or not.</p>

```python
# Importing itertools for looping/iteration.
import itertools 
# Numpy already imported in previous example, but we will use it again. Note, this notebook needs to be fully run,
# so numpy is present(imported).
```

```python
# Firstly, when we have our two adjacent matrices, we must check their degree sequeces are equal.
# We do this by arranging the vertex in non-increasing order. 
def degreeSequence(adjMat):
    # We make an empty array to store the result after every iteration.
    degSeq = []
    # Loop x times depending on size of matrix.
    for vertex in range(len(adjMat)):
        degSeq.append(sum(adjMat[vertex]))
    # Now we reverse sort the array and return it.
    degSeq.sort(reverse=True)
    return degSeq
```

```python
# Ensuring that the degree sequence is equal is just one of two crucial steps.
# We can easily determine if two adjacency matrices are not isomorphic again by checking 
# if the graph order of both graphs are not equal.

# We will use the below function to return the length of the matrix so we can compare the results
# of both graphs in the main function to help us further assess if they are isomorphic or not.
def graphOrder(adjMat):
    # If not, return -1 to ensure a wrong length.
    if len(adjMat) != len(adjMat[0]):
        return -1
    else:
        # return length
        return len(adjMat)
```

```python
# Here comes the "heavy" brute force part. This is terribly inefficient.
# We need to get ALL permutations of the passed matrix.

def getPermu(adjMat):
    # Oof, make an array for ALL permutations to be stored in.
    allMatrices = []
    idx = list(range(len(adjMat)))
    # Get all possible combinations.
    possCombo = [
        list(i) for i in itertools.permutations(idx, len(idx))
    ]
    # comb through combinations, transpose and add to array.
    for idx_comb in possCombo:
        a = adjMat
        a = a[idx_comb]
        # using NumpPy lib again here, to transpose.
        a = np.transpose(np.transpose(a)[idx_comb])
        # append to array.
        allMatrices.append({
            "perm_vertex":
            idx_comb,
            "adj_matrix":
            a
        })
    return allMatrices
    
```

```python
# Okay, we have all step necessary to check for two graphs are isomorphic or not. 
# Lets put all our functions to work inside this main function.

def bruteForseIso(adjMat1, adjMat2):
    degreeSeq1 = degreeSequence(adjMat1)
    degreeSeq2 = degreeSequence(adjMat2)
    # We now have the degree sequnce for both matrices.
    # As mentioned above, we have a second small step to check and save us a lot of computational power before
    # we even need to look at permutations, should they be flagged already as not isomorphic.
    if graphOrder(adjMat1) != graphOrder(adjMat2):
        # Not isomorphic, get out.
        return False
    # NumPy lib at work again.
    elif np.array_equal(degreeSeq1, degreeSeq2) == False:
        # Not isomorphic, get out.
        return False
    # Here it goes, first two check passed!
    else:
        for adj_matrix in list(
                map(lambda matrix: matrix["adj_matrix"],
                    # Get ALL permutations of the second matrix for comparison.
                    getPermu(adjMat2))):
            if np.array_equal(adjMat1, adj_matrix) == True:
                # We have a match!
                return True
    # Nothing hits, it is not isomorphic.
    return False    
```

# Computational complexity and runtime. 

***

<p>The above algoritm uses the brute force technique and as previsously mentioned this is a highly inefficient and costly technique. In order to determine if the graphs are isomorphic we must enumerate and test all permutations of the vertices of <i>graph2</i> until an isomorphic mapping is found. This assertion can only be made when every edge in <i>graph1</i> is also present in <i>graph2</i> using the currently enumerated permutation and vica versa. If, in the worst case, we compare all available permutations and we don't find an isomorphism, it concludes the algorithm as false and the two graphs are <b>not</b> isomorphic. With this information confirmed, we can say that this worst case scenario's run time is O($N^{2}2N!$) where N is the number of vertices in <i>graph1</i> and <i>graph2</i>. Due to the exponential growth that occurs when adding vertices, this algoritm would not scale well and the user would notice that their machines/devices would take too long to even run the algorithm at a certain point of adding another vertice.</p>
