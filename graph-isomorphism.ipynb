# P versus NP problem.

<p>To try to keep this straight forward initially, P is a set of relatively easy problems, and NP is a set of of what seems to be very hard problems. So, P = NP would imply that hard problems would actually have relatively easy solutions. But this has more complexity than that. So much that Clay Mathematics Institute will give you one million dollars for proving or disproving this theory!</p>

<p>The problem is unsolved in computer science. It creates the question if the solution of a problem can be quickly verified, can it also be solved quickly?</p>

<p>The term "quickly" is referred to the use of an algorithm, which provides a solution or outcome in polynomial time (<b>P</b>). A polynomial is an expression variables, terms, exponents and constants. Polynomial time is define by the running time is upper bounded by a polynomial expression which is the size of the input for the algorithm. Noted: T(n) = O(nk) for some positive constant k. For example, basic arithmetic operations (Addition, subtraction, etc) can be performed in polynomial time.</p>
<p>Sometimes, there is no known way to find an answer <b>but, if we are provided information showing the answer</b>, it is possible to verify the answer quickly. (<b>NP</b>) NP, which stands for nondeterministic polynomial, is the category these questions fall into if they can be answered/verified in polynomial time.</p>

<p>To conclude this section. If we are asked does P equal NP, we are asking if the solution to a problem can be verified in polynomial time, can it also be found in polynomial time.</p>

### A graph.

***

<p>First, let's evaluate what exactly a graph is. A graph itself can be thought of as a non-linear data structure. Comprised by a finite set of vertices which are connected by edges. The vertices represent entities in a graph and the edges represent the relationship between these entities, hence <i>Graph = (Vertices, Edges)</i>. Below I will use networkx for illstration purposes. Networkx is a python library used for studying graphs and networks.</p>

<b>Reference:  </b>[Networkx](https://networkx.org/documentation/stable/index.html)

```python
import networkx as nx
```

```python
# Prefine the parameters to instruct networkx on how we would like to draw the graph.
params = {'node_color': 'green', 'node_size' : 600, 'font_size': 16, 'with_labels': True}
```

```python
# First, create an empty graph.
graph = nx.Graph()
```

```python
# Now define a set of edges and how they are connected.
edges = ((0,1) , (1,2) , (0,2) , (0,3), (0, 4))
```

```python
# The nodes are created as needed, but we state to add the edges.
graph.add_edges_from(edges)
```

```python
# Now we can draw the graph. 
# Circular version.
nx.draw_circular(graph, **params)
```

```python
# Let's draw a couple versions of EXACTLY the same graph different ways using networkx.
# Spring.
nx.draw_spring(graph, **params)
```

```python
# Spring.
nx.draw_planar(graph, **params)
```

### Representing graphs.

***

    
<p>We can represent graphs by using three data structures:</p>
    
<ol>
    <li><b>Adjacency Matrix</b></li>
    <li><b>Adjacency list</b></li>
    <li><b>Adjacency set</b></li>
</ol>

<p>In this example, we will look at the adjacency matrix. We can think of an Adjaceny Matrix as a table with rows and columns. the rows and label columns represent the nodes of the graph. There must be equal amounts for rows and columns in order to be an adjacency matrix. Each field/entry in that matrix then represents the relationship between the two given nodes.</p>

<p>Let's take a look at four node graph, the relationships between them and represent their relationships using a matrix.</p>

```python
# Create an empty graph.
graph = nx.Graph()
```

```python
# Now define a set of edges and how they are connected.
edges = ((0, 1) , (1, 2) , (1, 3) , (2, 1), (3,2))
```

```python
# The nodes are created as needed, but we state to add the edges.
graph.add_edges_from(edges)
```

```python
# Now we can draw the graph. 
# Circular version.
nx.draw_circular(graph, **params)
```

### Apply the graph to a matrix.

***

<p>We have nodes 0 - 3. If we take a binary matrix, label our rows and colums 0 - 3 respectively. Each field in that matrix can be one of two values, 0 or 1 (binary), we can declare the presence of an edge between two nodes with 1 and the nonpresence by using 0.</p>

<p>We can efficiently create this matrix by using the networkx library but we will also use the numpy library to display matrices. This demonstrates how we can use graphs to demonstrate data structures such as this two dimensional array.</p>

<b>Reference:  </b>[NumPy Docs](https://numpy.org/doc/)

```python
# Importing the numpy library.
import numpy as np
```

```python
# This matrix represents the relationships in the previous graph.
print(nx.to_numpy_array(graph).astype(np.uint8))
```
